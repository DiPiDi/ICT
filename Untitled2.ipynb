```python
np.random.seed(123)

fertilizer = np.random.uniform(0, 200, 100)
crop_yield = 2 + 0.08 * fertilizer - 0.0001 * (fertilizer ** 2) + np.random.normal(0, 0.5, 100)
crop_yield[crop_yield < 0] = 0
data_fertilizer_df = pd.DataFrame({'Fertilizer (kg/hectare)': fertilizer, 'Crop Yield (tons/hectare)': crop_yield})

print("Number of null values per column:")
print(data_fertilizer_df.isnull().sum())

print("\nData cleaning (checking for nulls) complete.")
```

```python
num_data_points = len(data_fertilizer_df)
print(f"\nNumber of data points: {num_data_points}")

print("\nNumber of null values after potential cleaning:")
print(data_fertilizer_df.isnull().sum())

if data_fertilizer_df.isnull().sum().sum() == 0 and num_data_points >= 100:
    print("\nData is clean (no null values) and contains at least 100 data points.")
else:
    print("\nData cleaning or data point count needs attention.")
```

```python
fertilizer_np = data_fertilizer_df['Fertilizer (kg/hectare)'].to_numpy()
crop_yield_np = data_fertilizer_df['Crop Yield (tons/hectare)'].to_numpy()

print("\nNumPy Operations:")

mean_fertilizer = np.mean(fertilizer_np)
print(f"   Mean fertilizer: {mean_fertilizer:.2f}")

max_yield = np.max(crop_yield_np)
print(f"   Maximum crop yield: {max_yield:.2f}")

fertilizer_range = np.ptp(fertilizer_np)
print(f"   Range of fertilizer: {fertilizer_range:.2f}")

std_yield = np.std(crop_yield_np)
print(f"   Standard deviation of crop yield: {std_yield:.2f}")

argmax_yield = np.argmax(crop_yield_np)
print(f"   Index of maximum crop yield: {argmax_yield}")
```
