```python
import pandas as pd      
import numpy as np       
import matplotlib.pyplot as plt   
import seaborn as sns    
from scipy import stats  
import statsmodels.api as sm 
```

```python
np.random.seed(42)  

data = {
    'Crop_Yield': np.random.normal(5, 1.5, 150),           
    'Fertilizer_Amount': np.random.uniform(50, 200, 150),  
    'Rainfall': np.random.uniform(100, 300, 150),         
    'Temperature': np.random.uniform(15, 30, 150)         
}

df = pd.DataFrame(data)

print("Here is the start of our data:")
print(df.head())
```

```python
print("\nCheck for missing values in each column:")
print(df.isnull().sum()) 

df_cleaned = df.dropna()

print("\nNumber of data points after cleaning:", len(df_cleaned))
```

```python
mean_yield = np.mean(df_cleaned['Crop_Yield'])
print("Average Crop Yield:", mean_yield)

std_fertilizer = np.std(df_cleaned['Fertilizer_Amount'])
print("Fertilizer Amount Standard Deviation:", std_fertilizer)

correlation = np.corrcoef(df_cleaned['Crop_Yield'], df_cleaned['Fertilizer_Amount'])[0, 1]
print("Correlation between Crop Yield and Fertilizer Amount:", correlation)

max_rainfall = np.max(df_cleaned['Rainfall'])
print("Maximum Rainfall:", max_rainfall)

sum_temperature = np.sum(df_cleaned['Temperature'])
print("Sum of Temperatures:", sum_temperature)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df_cleaned['Fertilizer_Amount'], df_cleaned['Crop_Yield'])

print("Linear Regression Results:")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared value:", r_value**2) 
```

```python
X = sm.add_constant(df_cleaned['Fertilizer_Amount'])  
y = df_cleaned['Crop_Yield']

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
plt.figure(figsize=(8,5)) 
sns.scatterplot(x='Fertilizer_Amount', y='Crop_Yield', data=df_cleaned) 
plt.title('Crop Yield vs Fertilizer Amount') 
plt.xlabel('Fertilizer Amount (kg/ha)') 
plt.ylabel('Crop Yield (tons/ha)')  # Y-axis label
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(8,5))  # Set the size of the plot
sns.histplot(df_cleaned['Crop_Yield'], bins=20, kde=True)  # Create the histogram
plt.title('Distribution of Crop Yield')  # Title of the plot
plt.xlabel('Crop Yield (tons/ha)')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(8,5))  # Set the size of the plot
sns.regplot(x='Fertilizer_Amount', y='Crop_Yield', data=df_cleaned, scatter_kws={'alpha':0.5})  # Create the regression plot
plt.title('Correlation and Regression between Fertilizer Amount and Crop Yield')  # Title of the plot
plt.xlabel('Fertilizer Amount (kg/ha)')  # X-axis label
plt.ylabel('Crop Yield (tons/ha)')  # Y-axis label
plt.show()  # Display the plot
```
